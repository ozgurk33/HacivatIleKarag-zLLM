```python
from pytube import Playlist
import yt_dlp
import time
import random
import os
import torch
import whisper
import gc
import json

video_urls = []

def get_playlist_url(playlist_url, number_of_urls):
    playlist = Playlist(playlist_url)
    for i, video in enumerate(playlist.video_urls):
        if i >= number_of_urls:
            break
        if i == 26 or i == 33:
            continue
        video_urls.append(video)

def download_youtube_mp3(video_urls, output_folder="audios", wait_range=(10,17)):
    ydl_opts = {
        'format': 'bestaudio/best',
        'outtmpl': f'{output_folder}/%(title)s.%(ext)s',
        'postprocessors': [{
            'key': 'FFmpegExtractAudio',
            'preferredcodec': 'mp3',
            'preferredquality': '192',
        }],
    }
    os.makedirs(output_folder, exist_ok=True)
    with yt_dlp.YoutubeDL(ydl_opts) as ydl:
        for url in video_urls:
            try:
                print(f"🎵 İndiriliyor: {url}")
                ydl.download([url])
                wait_time = random.randint(wait_range[0], wait_range[1])
                print(f"⏳ {wait_time} saniye bekleniyor...")
                time.sleep(wait_time)
            except Exception as e:
                print(f"⚠️ Hata oluştu: {url} -> {e}")

AUDIO_FOLDER = "/content/drive/MyDrive/audios"
OUTPUT_FOLDER = "/content/drive/MyDrive/Transcripts"
os.makedirs(OUTPUT_FOLDER, exist_ok=True)

device = "cuda" if torch.cuda.is_available() else "cpu"
print(f"Kullanılan cihaz: {device}")

print("Whisper modeli yükleniyor (large-v3)...")
model = whisper.load_model("large-v3", device=device)
print("Whisper modeli yüklendi!")

for file_name in os.listdir(AUDIO_FOLDER):
    if file_name.endswith(".mp3"):
        audio_path = os.path.join(AUDIO_FOLDER, file_name)
        print(f"\nTranscribe ediliyor: {file_name}")

        result = model.transcribe(audio_path, language="tr", fp16=(device=="cuda"))

        text_output = os.path.join(
            OUTPUT_FOLDER, f"{os.path.splitext(file_name)[0]}_transcript.txt"
        )
        with open(text_output, "w", encoding="utf-8") as f:
            f.write(result["text"])

        segments_output = os.path.join(
            OUTPUT_FOLDER, f"{os.path.splitext(file_name)[0]}_segments.json"
        )
        with open(segments_output, "w", encoding="utf-8") as f:
            json.dump(result["segments"], f, ensure_ascii=False, indent=2)

        print(f"Transkripsiyon ve segmentler kaydedildi: {text_output}, {segments_output}")

del model
gc.collect()
torch.cuda.empty_cache()
print("\n💾 Tüm MP3 dosyaları metne dönüştürüldü ve zaman bilgileri kaydedildi.")
```

```python
!pip install pyannote.audio --quiet

import os
import json
from pyannote.audio import Pipeline

AUDIO_FOLDER = "/content/drive/MyDrive/audios"
INPUT_FOLDER = "/content/drive/MyDrive/Transcripts"
OUTPUT_FOLDER = "/content/drive/MyDrive/Merged_Transcripts"
os.makedirs(OUTPUT_FOLDER, exist_ok=True)

hf_token = ""

print("Pyannote speaker diarization modeli yükleniyor...")
pipeline = Pipeline.from_pretrained("pyannote/speaker-diarization", use_auth_token=hf_token)
print("Model yüklendi!")

speaker_map = {"SPEAKER_00": "Hacivat", "SPEAKER_01": "Karagöz"}

def intersect(a_start, a_end, b_start, b_end):
    return not (a_end < b_start or a_start > b_end)

def load_json(path):
    with open(path, "r", encoding="utf-8") as f:
        return json.load(f)

for file_name in os.listdir(INPUT_FOLDER):
    if not file_name.endswith("_segments.json"):
        continue

    base = file_name.replace("_segments.json", "")
    audio_path = os.path.join(AUDIO_FOLDER, f"{base}.mp3")
    whisper_path = os.path.join(INPUT_FOLDER, file_name)

    if not os.path.exists(audio_path):
        print(f"Ses dosyası bulunamadı: {audio_path}")
        continue

    print(f"\nİşleniyor: {base}")

    diarization = pipeline(audio_path)
    diar_segments = [
        {"start": round(turn.start, 2), "end": round(turn.end, 2), "speaker": speaker}
        for turn, _, speaker in diarization.itertracks(yield_label=True)
    ]

    whisper_segments = load_json(whisper_path)

    merged = []
    for seg in whisper_segments:
        seg_start, seg_end = seg["start"], seg["end"]
        matched_speaker = "Bilinmiyor"
        for d in diar_segments:
            if intersect(seg_start, seg_end, d["start"], d["end"]):
                matched_speaker = d["speaker"]
                break
        merged.append({
            "start": seg_start,
            "end": seg_end,
            "speaker": speaker_map.get(matched_speaker, matched_speaker),
            "text": seg["text"].strip()
        })

    merged_json = os.path.join(OUTPUT_FOLDER, f"{base}_merged.json")
    merged_txt = os.path.join(OUTPUT_FOLDER, f"{base}_dialog.txt")

    with open(merged_json, "w", encoding="utf-8") as f:
        json.dump(merged, f, ensure_ascii=False, indent=2)

    with open(merged_txt, "w", encoding="utf-8") as f:
        for m in merged:
            f.write(f"{m['speaker']}: {m['text']}\n")

    print(f"{base}: Tamamlandı → {merged_txt}")

print("\n🎉 Tüm dosyalar işlendi! Çıktılar klasörü:")
print(OUTPUT_FOLDER)
```
